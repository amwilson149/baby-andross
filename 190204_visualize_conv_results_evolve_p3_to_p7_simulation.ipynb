```python
import numpy as np
import json
import matplotlib.pyplot as plt
```

Set figure font size

```python
FS = 65
```

Get results for convergence from files

```python
fname = './data/p3_p7_evolve_results/190206_evolve_p3_p7_coarse_scan_pr_0_1.0_g_-1.5_2.5_avg_trial_conv_results.json' # coarse first pass
fnamef = './data/p3_p7_evolve_results/190208_evolve_p3_p7_fine_scan_pr_0.0_0.1_g_0.0_2.0_avg_trial_conv_results.json' # fine second pass
```

```python
with open(fname,'r') as f:
    conv_results = json.loads(f.read())
with open(fnamef,'r') as f:
    convf_results = json.loads(f.read())
```

Inspect headings

```python
conv_results.keys()
```

```python
type(conv_results['gamma'])
```

```python
ptsize = 180

plt.figure(figsize=(20,20))
plt.scatter(conv_results['prem'],conv_results['gamma'],s=[ptsize for q in range(len(conv_results['gamma']))],c = conv_results['frac_conv_runs'])
plt.xlabel('Relative synapse removal rate',fontsize=FS)
plt.ylabel('$\gamma$',fontsize=FS)
plt.tick_params(labelsize=FS)
cbar = plt.colorbar()
cbar.ax.tick_params(labelsize=FS)
# cbar.set_label('Fraction of convergent runs',fontsize=FS,rotation=270)
fname = './data/figures/p3_p7_evolve/190213_conv_frac_coarse_pass.png'
# plt.show()
plt.savefig(fname)
```

```python
ptsize = 160

plt.figure(figsize=(20,20))
plt.scatter(convf_results['prem'],convf_results['gamma'],s=[ptsize for q in range(len(convf_results['gamma']))],c = convf_results['frac_conv_runs'])
plt.xlabel('Relative synapse removal rate',fontsize=FS)
plt.ylabel('$\gamma$',fontsize=FS)
plt.tick_params(labelsize=FS)
cbar = plt.colorbar()
cbar.ax.tick_params(labelsize=FS)
# cbar.set_label('Fraction of convfergent runs',fontsize=FS,rotation=270)
fname = './data/figures/p3_p7_evolve/190213_conv_frac_fine_pass.png'
# plt.show()
plt.savefig(fname)
```
